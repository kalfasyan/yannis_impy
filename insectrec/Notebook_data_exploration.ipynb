```python
%reset -f
import pandas as pd
import numpy as np
import os, glob
```

```python
df_imgs = pd.DataFrame(os.listdir('./saved_imgs/'), columns=['fnames'])
df_imgs['labels'] = df_imgs['fnames'].apply(lambda x: x.split('.')[0])

dict_or_df = 'dict'

BASE_DATA_DIR = "/home/kalfasyan/data/insects/Sticky_Plates"
SAVE_DIR = "./saved_imgs/"
week_folders = ['Week 30 31 32', 'week 36 37 38', 'Week  33 34 35', 'Week 26  28 29 30']

xlsx_files = [fname for fname in glob.iglob(BASE_DATA_DIR + '/**/*.xlsx', recursive=True)]
xlsx_classes = [fname for fname in xlsx_files if fname.split('/')[-1].split('.')[-2].endswith('geklasseerd')]
dataframes = []
for f in xlsx_classes:
    sub = pd.read_excel(f)
    sub['ID'] = sub['name plate'].map(str) + '_' + sub['index'].map(str) # '_' + sub['Klasse'].map(str) + 
    dataframes.append(sub[['ID','Klasse']])
df = pd.concat(dataframes, axis=0)
if dict_or_df == 'dict':
    tmp = pd.Series(df['Klasse'].values, index=df['ID']).to_dict()
elif dict_or_df == 'df':
    tmp = df
else:
    raise ValueError('Didnt specify what to return correctly!')
    
df_imgs.replace({"labels": tmp}, inplace=True)
classes_to_keep = df_imgs.sort_values(by='fnames')['labels'].value_counts().iloc[:9].index.tolist()
plate = df_imgs[df_imgs.labels=='wmv'].fnames.iloc[3]
df_imgs[df_imgs.fnames.apply(lambda x: x[:-6]) == plate[:-6]]
```

```python
df['index'] = df['ID'].apply(lambda x: int(x.split('_')[-1]))
df['fnames'] = df['ID'].apply(lambda x: '_'.join(x.split('_')[:-1]))
df['class'] = df['Klasse']
sub = df[['fnames','index','class']]
sub.to_csv('class_mapping.csv', index=False)
```
